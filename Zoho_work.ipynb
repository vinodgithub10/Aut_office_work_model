```python
import requests
import pandas as pd

# List of IDs
ids = [
  "1661984000007962053"

]

# Empty lists to store extracted data
id_list = []
contact_names = []
billing_address_ids = []
shipping_address_ids = []

# Loop through each ID and make API requests
for i in ids:
    url = f"https://iapi.getpitstop.com/v1.0/pitstopspare/zoho/v1/inventory/contact/detail/ACPL/{i}"
    response = requests.get(url)
    if response.status_code == 200:
        json_data = response.json()
        
        # Extracting Contact Name
        contact_name = json_data['data']['contactDetail']['contact']['contactName']

        # Extracting billingAddress addressId
        billing_address_id = json_data['data']['contactDetail']['contact']['billingAddress']['addressId']

        # Extracting shippingAddress addressId
        shipping_address_id = json_data['data']['contactDetail']['contact']['shippingAddress']['addressId']

        # Append the extracted data to the lists
        id_list.append(i)
        contact_names.append(contact_name)
        billing_address_ids.append(billing_address_id)
        shipping_address_ids.append(shipping_address_id)
    else:
        print(f"Failed to fetch data for ID: {i}")

# Creating a DataFrame to store the extracted data
data = {
    "ID": id_list,
    "Contact Name": contact_names,
    "Billing Address ID": billing_address_ids,
    "Shipping Address ID": shipping_address_ids
}

# Converting the data to a Pandas DataFrame
df = pd.DataFrame(data)

df
```
